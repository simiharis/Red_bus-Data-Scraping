```python
# importing libraries
import mysql.connector
import pandas as pd
import numpy as np
```

```python
# converting csv files into dataframes
df_A = pd.read_csv("APSRTC_bus_details.csv")
df_T = pd.read_csv("TSRTC_bus_details.csv")
df_U = pd.read_csv("UPSRTC_bus_details.csv")
df_SB = pd.read_csv("South_Bengal_bus_details.csv")
df_H = pd.read_csv("HRTC_bus_details.csv")
df_W = pd.read_csv("WBTC_bus_details.csv")
df_R = pd.read_csv("RSRTC_bus_details.csv")
df_K = pd.read_csv("KSRTC_bus_details.csv")
df_BS = pd.read_csv("BSRTC_bus_detais.csv")
df_AS=pd.read_csv("ASTC_bus_detais.csv")
```

```python
# Combining the individual data frames to a single data frame
combined_df=pd.concat([df_A,df_T,df_U,df_SB,df_H,df_W,df_R,df_K,df_BS,df_AS],ignore_index=True)
combined_df
```

DATA CLEANING

```python
combined_df.info() # to get the information of the data frame
```

```python
combined_df.drop_duplicates(inplace=True) # to remove the duplicates in the data frame
```

```python
combined_df.isnull().sum() # to get the sum of null values in the data frame
```

```python
# Convert the Price column to string to handle mixed data types
combined_df['Price'] = combined_df['Price'].astype(str)

# Remove 'INR' from the Price column and strip any extra spaces
combined_df['Price'] = combined_df['Price'].str.replace('INR', '').str.strip()

# Convert the cleaned column back to numeric (float)
combined_df['Price'] = pd.to_numeric(combined_df['Price'], errors='coerce')
combined_df['Price'].fillna=0
```

```python
combined_df.info()
```

```python
combined_df_= combined_df[combined_df["Price"] <= 7000]
```

```python
# replacing the nan value
combined_df =combined_df.replace({np.nan: None})
```

```python
#sql connection
conn = mysql.connector.connect(host="localhost", user="root", password="simi")
my_cursor = conn.cursor()
my_cursor.execute("CREATE DATABASE IF NOT EXISTS REDBUSINFO")
conn.database = 'REDBUSINFO'
```

```python
# Table Creation
my_cursor.execute('''CREATE TABLE IF NOT EXISTS busdata1(
                  ID INT AUTO_INCREMENT PRIMARY KEY,
                  Route_name VARCHAR(250) NULL,
                  Route_link VARCHAR(250) NULL,
                  Bus_name VARCHAR(250) NOT NULL,
                  Bus_type VARCHAR(250) NOT NULL,
                  Departing_time VARCHAR(250) NOT NULL,
                  duration VARCHAR(250) NOT NULL,
                  Reaching_time VARCHAR(250) NOT NULL,
                  Star_Rating FLOAT NULL,
                  Price Float NOT NULL,
                  Seat_Availability VARCHAR(250) NOT NULL
                  )''')
print("Table Created successfully")
```

```python
# SQL query to insert data into Bus_data table
insert_query = '''INSERT INTO busdata1(
                    Route_name,
                    Route_link,
                    Bus_name,
                    Bus_type,
                    Departing_time,
                    duration,
                    Reaching_time,
                    Star_Rating,
                    Price,
                    Seat_Availability
                )
                VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)'''

# Convert DataFrame to list of lists
data = combined_df.values.tolist()





try:
    my_cursor.executemany(insert_query, data)
    conn.commit()
    print("Data inserted successfully.")
except Exception as e:
    conn.rollback()  # Roll back if there is any error
    print(f"An error occurred: {e}")



print("Values inserted successfully")
```
